```python
def outlierDetection(column):
    #Sort in Ascending
    sorted(column)
    #Get Q3 and Q1
    Q1,Q3 = np.percentile(column, [25,75])
    #Calc IQR
    IQR = Q3 - Q1
    #Calc lowerRange
    lowerRange = Q1 - (1.5 * IQR)
    #Calc upperRange
    upperRange = Q3 + (1.5 * IQR)
    #Return Ranges
    return lowerRange,upperRange
```

# Regression model for seprate store

```python
 print('REGRESSION FOR SEPRATE STORE(WEEK_1)')
for i in range(1,8):
    store = train_data.groupby('Store')
    storei = store.get_group(i)
    x_train_storei = storei.iloc[:,[5,8,12]].values
    y_train_storei = storei.iloc[:,[10]].values

    from sklearn.linear_model import LinearRegression
    model_storei = LinearRegression()
    model_storei.fit(x_train_storei ,y_train_storei )
    train_score=model_storei.score(x_train_storei ,y_train_storei)
    train_score

    hidden_storei=hidden_data.groupby("Store").get_group(i)
    #hidden_storei.head(2)

    x_hidden = hidden_storei.iloc[:,[5,8,12]].values
    y_hidden = hidden_storei.iloc[:,[10]].values

    model_storei.fit(x_hidden ,y_hidden )
    test_score=model_storei.score(x_hidden ,y_hidden)
    test_score

    test_storei=test_data.groupby("Store").get_group(i)
    #test_storei.head()

    x_test = test_storei.iloc[:,[5,8,11]].values
    y_pred = model_storei.predict(x_test)


    mae=mean_absolute_error(y_pred, y_hidden)
    mse=mean_squared_error(y_pred, y_hidden)
    from math import sqrt

    def rmse(predictions, targets):
        return np.sqrt(((y_pred - y_hidden) ** 2).mean())
    rmse=rmse(y_pred,y_hidden)
    #print('                                     REGRESSION FOR SEPRATE STORE(WEEK_1)')
    print('STORE{},  MSE {},            MAE {},    RMSE {} '.format(i,mse,mae,rmse))
    print('         train_score {}'.format(train_score))
    print('         test_score  {}'.format(test_score))
   
```

```python
def outlierDetection(column):
    #Sort in Ascending
    sorted(column)
    #Get Q3 and Q1
    Q1,Q3 = np.percentile(column, [25,75])
    #Calc IQR
    IQR = Q3 - Q1
    #Calc lowerRange
    lowerRange = Q1 - (1.5 * IQR)
    #Calc upperRange
    upperRange = Q3 + (1.5 * IQR)
    #Return Ranges
    return lowerRange,upperRange
```

```python
lr,ur = outlierDetection(data.Customers)
lr,ur
```

```python
data.drop(data.Customers[(data.Customers > ur) | (data.Customers < lr)].index , inplace=True )
sns.distplot(data.Customers)
```

```python
lr,ur = outlierDetection(data.Sales)
lr,ur
```

```python
data.drop(data.Sales[(data.Sales > ur) | (data.Sales < lr)].index , inplace=True )
sns.distplot(data.Sales)
```
